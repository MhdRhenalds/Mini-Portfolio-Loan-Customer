## Handling Outlier

```python
numeric_columns = ['applicant_income','coapplicant_income', 'loan_amount', 'loan_term_year', 'credit_history']

plt.figure(figsize=(15, 6))
sns.boxplot(data=df[numeric_columns], orient='h')
plt.title('Boxplot of Numeric Colums')
plt.xticks(rotation=45)
plt.show()
```

```python
cols_to_check = ['applicant_income','coapplicant_income', 'loan_amount', 'loan_term_year', 'credit_history']

for col in cols_to_check:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    low_limit = Q1 - 1.5 * IQR
    high_limit = Q3 +1.5 * IQR

    df = df[(df[col] >= low_limit) & (df[col] <= high_limit)]
```

## Analisis pengaruh status perkawinan terhadap jangka waktu meminjam 

```python
df_gr2 = df.groupby(['married', 'loan_term_year'])['loan_id'].count().reset_index()
df_gr2
```

```python
#Mengkateogrikan berdasarkan loan_term_year

loan_term_year_group = {}
for i in range(1,41):
    if i<15:
        loan_term_year_group[i] = '<15'
    elif i>=15 and i<=25:
        loan_term_year_group[i] = '15-25'
    elif i==30:
        loan_term_year_group[i] = '>=30'
    else:
        pass

#membuat kolom baru berdasarkan pada pengelompokan sebelumnya
df_gr2['loan_term_year_group'] = df_gr2['loan_term_year'].replace(loan_term_year_group)
df_gr2
```

```python
plt.figure(figsize=(12,6)) #memperkecil ukuran plot
sns.barplot(x='loan_term_year_group', y='loan_id', hue='married', data=df_gr2)

plt.tick_params(axis='both', which='major', labelsize=10) #Memperkecil ukuran
plt.grid()
plt.legend(title='Status perkawinan', title_fontsize=12, prop={'size': 10}, loc='upper right') #Meneysuaikan legend

plt.xlabel('Jangka waktu pembiayaan (dalam tahun)', fontsize=12)
plt.ylabel('Jumlah pemohon pembiayaan', fontsize=12)
plt.title('Nasabah yang telah menikah lebih banyak mengajukan pembiayaan dengan jangka waktu 30 tahun ke atas', fontsize=14, fontweight='bold') #mengurangi ukuran judul
plt.tight_layout() #menyesuaikan tata letak plot
plt.show();

```
